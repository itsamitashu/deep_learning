```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sentence=["This", "movie", "is", "very", "good"]

weights=torch.tensor([0.05, 0.15, 0.1, 0.3, 0.4])
weights=F.softmax(weights, dim=0).numpy()
```

```python
sns.heatmap([weights], annot=True, xticklabels=sentence)
plt.title("Attention Heatmap")
plt.xlabel("Words")
plt.show()
```
